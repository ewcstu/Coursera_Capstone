```python
#url =  'https://en.wikipedia.org/wiki/List_of_city-designated_neighbourhoods_in_Toronto'
url =  'https://en.wikipedia.org/wiki/Demographics_of_Toronto_neighbourhoods'

raw_page = requests.get(url)
page = raw_page.text
print(page)
```

```python
df = pd.read_html(page)
df = df[1] # 0 for url1, 1 for url2

# Clean rows/columns
df1 = df[df.Name != "Toronto CMA Average"]
df2 = df1[['Name','Population','Land area (km2)','Average Income']]
df2.head()
```

```python
# name it DF and easier column names
df = df2
df.columns = ['Neighborhood','Pop','Area','Income']
df.head()
```

```python
# geopy to get lat/long - test on toronto
address = 'Toronto, CN'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print ('Toronto coordinates: {}, {}'.format(latitude, longitude))
```

```python
# test geocode on another city
address = 'Agincourt, Toronto, Canada'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print ('Coordinates: {}, {}'.format(latitude, longitude))
```

```python
# geopy to get lat/long
address = df['Neighborhood'] + ', Toronto, Canada'
df['Coordinates'] = address.apply(geolocator.geocode)
df.head()
```
