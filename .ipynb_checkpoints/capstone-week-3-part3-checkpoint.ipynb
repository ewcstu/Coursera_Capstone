### Part 1

```python
# imports
import pandas as pd
import numpy as np
import requests
from geopy.geocoders import Nominatim
from bs4 import BeautifulSoup
```

1. Using the postal code list from this page:
https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, 
reading the website, then loading for beautifulsoup to parse

```python
url =  'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

page = requests.get(url).text
soup = BeautifulSoup(page,'xml')
```

2. Using beautifulsoup, extract the cells we want - all rows where the Borough is not assigned - and for rows when the Neighborhood is not assigned, giving it the Borough name. Then loading into a pandas dataframe

```python
table = soup.find('table')
table_rows = table.find_all('tr')
rows = []
for tr in table_rows:
    r = tr.find_all('td')
    if r:
        rows.append(r)
        
data = []
for row in rows:
    a = row[0].text.strip()
    b = row[1].text.strip()
    c = row[2].text.strip()
    if b != 'Not assigned':
        if c == 'Not assigned':
            c = b
        data.append([a, b, c])

df = pd.DataFrame(data,columns=['PostalCode','Borough','Neighborhood'])
#df.head()
```

3. Now we have a row for every Neighborhood, but we want to use groupby so that each postal code only has one row.

```python
df_2 = df.groupby(by = ['PostalCode','Borough']).agg(lambda x: ','.join(x))
df_2.reset_index(level=['PostalCode','Borough'], inplace=True)
#df_2.head()
```

4. Using the .shape method to show the # of rows in the dataframe (103 rows):

```python
print(df_2.shape)
```

### Part 2

```python
coords = pd.read_csv('Geospatial_Coordinates.csv')
coords.head()
print(coords.shape)
```

2. Join the two tables into an output dataframe providing PostalCode, Borough, list of Neighborhoods, Latitude, an|d Longitudes. Check the head to make sure it's formatted correctly.

```python
joined = df_2.set_index('PostalCode').join(coords.set_index('Postal Code'))
output = joined.reset_index()
output.head()
```

```python
# mapping imports
import folium
# print ('Folium imported')
```

1. Decide where to center the map - find the coordinates of Toronto

```python
# Find coordinates for Toronto
geolocator = Nominatim(user_agent="to_locator")
address = 'Toronto'
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of {} are {}, {}.'.format(address, latitude, longitude))
```

2. For ease, make a smaller dataframe of just the fields we want to show

```python
mapdf = output[["PostalCode","Latitude","Longitude"]]
```

3. Create the world map centered on Toronto (coordinates just found) and create a marker for each postal code based on the dataframe from part 2.

```python
world_map = folium.Map(location=[43.653963,-79.387207])

for a, b, c in zip(mapdf['Latitude'], mapdf['Longitude'], mapdf['PostalCode']):
    popup = folium.Popup(c)
    folium.CircleMarker(
        [a, b],
        popup=popup
        ).add_to(world_map)
world_map
```
